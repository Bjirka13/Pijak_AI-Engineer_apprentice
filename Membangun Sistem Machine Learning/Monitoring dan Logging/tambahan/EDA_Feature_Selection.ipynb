Distance Analysis

```python
from math import radians, cos, sin, asin, sqrt
import numpy as np

def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # km
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)

    a = sin(dlat/2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))

    return R * c

df['distance_proxy'] = np.sqrt(df['latitude']**2 + df['longitude']**2)
```

```python
df.groupby(pd.qcut(df['distance_proxy'], 5))['late_delivery_risk'].mean()
```

```python
# =========================================================
# QUANTILE THRESHOLD VALIDATION
# =========================================================

quantiles = [i/100 for i in range(10, 91, 10)]
results = []

for q in quantiles:
    threshold = (
        df["sales"]
        .quantile(q)
    )

    temp_flag = (
        df["sales"] >= threshold
    ).astype(int)

    late_rate = (
        df.groupby(temp_flag)
        ["late_delivery_risk"]
        .mean()
    )

    gap = abs(
        late_rate[1] - late_rate[0]
    )

    results.append({
        "quantile": q,
        "threshold": threshold,
        "non_high_value_risk":
            late_rate[0],

        "high_value_risk":
            late_rate[1],

        "risk_gap": gap
    })

quantile_validation_df = (
    pd.DataFrame(results)
)

display(
    quantile_validation_df
)
```
